```python
import numpy as np
import pandas as pd
import scipy.stats as st
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def calculate_stats(a,b,a_orig,b_orig,n):
    count=0
    orig_diff=abs(a_orig-b_orig)
    
    diff=[i-j for i,j in zip(a,b)]
    print(orig_diff)
    
    mean=np.mean(diff)
    print(diff[:10])
    print(diff[:10]-mean)
    print(mean)
    
    for i in range(0,n):
        
        if abs(diff[i]-mean)>=orig_diff:
            count+=1
    
    p_value=float(count)/float(n)
    #print(count)
    print('p_value',p_value)
    if p_value<=0.05:
        
        print('false')
    else:
        
        print('true')
```

```python
def stats_plot(a,b,label1,label2):
    
    print(np.mean(a))

    print(np.percentile(a,2.5))
    print(np.percentile(a,97.5))
    
    print(np.mean(b))

    print(np.percentile(b,2.5))
    print(np.percentile(b,97.5))
    

    print(st.t.interval(0.95, len(a)-1, loc=np.mean(a), scale=st.sem(a)))
    print(st.norm.interval(0.05, loc=np.mean(a), scale=st.sem(a)))

    ci = norm(*norm.fit(a)).interval(0.95)  # fit a normal distribution and get 95% c.i.
    height, bins, patches = plt.hist(a,color='red',alpha=0.3,label=label1)
    plt.fill_betweenx([0, height.max()], ci[0], ci[1], color='g', alpha=0.1)

    ci = norm(*norm.fit(b)).interval(0.95)  # fit a normal distribution and get 95% c.i.
    height, bins, patches = plt.hist(b,color='green',alpha=0.3,label=label2)
    plt.fill_betweenx([0, height.max()], ci[0], ci[1], color='g', alpha=0.1)
    #plt.hist(scores_bleu)
    plt.xlabel("BLEU Scores")
    plt.ylabel("#Samples")
    plt.legend()
    #plt.savefig('bootstrap_stats_baseline3_SAT1.pdf',dpi=1000)
   
    plt.show()
```
